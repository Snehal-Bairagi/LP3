```python
class Item:
    def __init__(self, value, weight):
        self.value = value
        self.weight = weight
        self.ratio = value / weight  # value-to-weight ratio

def fractional_knapsack(capacity, items):
    # Sort items by value/weight ratio (descending)
    items.sort(key=lambda x: x.ratio, reverse=True)
    total_value = 0
    selected = []

    for item in items:
        if capacity == 0:
            break
        if item.weight <= capacity:
            # Take the full item
            capacity -= item.weight
            total_value += item.value
            selected.append((item.value, item.weight, 1.0))
        else:
            # Take a fraction of the item
            fraction = capacity / item.weight
            total_value += item.value * fraction
            selected.append((item.value, item.weight, fraction))
            capacity = 0

    return total_value, selected


# -------------------------------
# Take input from user
# -------------------------------
n = int(input("Enter number of items: "))
items = []

for i in range(n):
    print(f"\nEnter details for Item {i+1}:")
    value = float(input("Value: "))
    weight = float(input("Weight: "))
    items.append(Item(value, weight))

capacity = float(input("\nEnter knapsack capacity: "))

# -------------------------------
# Solve and display output
# -------------------------------
max_value, selected_items = fractional_knapsack(capacity, items)

print("\n--- Fractional Knapsack Result ---")
print("Selected Items (Value, Weight, Fraction Taken):")
for val, wt, frac in selected_items:
    print(f"Value: {val}, Weight: {wt}, Fraction taken: {frac:.2f}")

print(f"\nTotal Maximum Value: {max_value:.2f}")
```
