```python
# Recursive Fibonacci with steps
steps_recursive = 0
def fib_recursive(n):
    global steps_recursive
    steps_recursive += 1
    if n <= 1:
        return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)

# Iterative Fibonacci with steps
def fib_iterative(n):
    steps = 0
    a, b = 0, 1
    if n == 0:
        steps += 1
        return 0, steps
    elif n == 1:
        steps += 1
        return 1, steps
    for _ in range(2, n + 1):
        steps += 1
        a, b = b, a + b
    return b, steps

# Take input from user
n = int(input("Enter the value of n: "))

# Iterative method
iter_val, iter_steps = fib_iterative(n)
print("\n--- Iterative Method ---")
print("Fibonacci Number:", iter_val)
print("Step Count:", iter_steps)

# Recursive method
steps_recursive = 0
rec_val = fib_recursive(n)
print("\n--- Recursive Method ---")
print("Fibonacci Number:", rec_val)
print("Step Count:", steps_recursive)
```
